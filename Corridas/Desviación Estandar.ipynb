```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import ascii
import matplotlib.colors as colors
from matplotlib import *
import sys
from pylab import *
from matplotlib.colors import LogNorm
import scipy.integrate as integrate
import scipy
from astropy.io import fits
import math
import os,glob
import matplotlib.patches as mpatches
from scipy.interpolate import splrep, splev
import numpy.polynomial.polynomial as poly



N=27000
mylist4  = []
mylist8  = []
mylist16  = []
mylist32  = []
mylist64  = []
mylist128  = []

p4=np.zeros((N,), dtype=np.float)
p8=np.zeros((N,), dtype=np.float)
p16=np.zeros((N,), dtype=np.float)
p32=np.zeros((N,), dtype=np.float)
p64=np.zeros((N,), dtype=np.float)
p128=np.zeros((N,), dtype=np.float)

with open('/home/pedro/Escritorio/Fisica-Computacional-master/Corridas/dispersionA_L=4.dat', 'r') as Gf:
    for line in Gf:
        mylist4.append(line)
Gf.close()

for i in range(N): 
        lista = mylist4[i].split()
        p4[i] = float(lista[0])

with open('/home/pedro/Escritorio/Fisica-Computacional-master/Corridas/dispersionA_L=8.dat', 'r') as Gf:
    for line in Gf:
        mylist8.append(line)
Gf.close()

for i in range(N): 
        lista = mylist8[i].split()
        p8[i] = float(lista[0])
with open('/home/pedro/Escritorio/Fisica-Computacional-master/Corridas/dispersionA_L=16.dat', 'r') as Gf:
    for line in Gf:
        mylist16.append(line)
Gf.close()

for i in range(N): 
        lista = mylist16[i].split()
        p16[i] = float(lista[0])

with open('/home/pedro/Escritorio/Fisica-Computacional-master/Corridas/dispersionA_L=32.dat', 'r') as Gf:
    for line in Gf:
        mylist32.append(line)
Gf.close()

for i in range(N): 
        lista = mylist32[i].split()
        p32[i] = float(lista[0])

with open('/home/pedro/Escritorio/Fisica-Computacional-master/Corridas/persionA_L=64.dat', 'r') as Gf:
    for line in Gf:
        mylist64.append(line)
Gf.close()

for i in range(N): 
        lista = mylist64[i].split()
        p64[i] = float(lista[0])
        
with open('/home/pedro/Escritorio/Fisica-Computacional-master/Corridas/dispersionA_L=128.dat', 'r') as Gf:
    for line in Gf:
        mylist128.append(line)
Gf.close()

for i in range(N): 
        lista = mylist128[i].split()
        p128[i] = float(lista[0])   
        
x = np.array([(np.std(p4)),(np.std(p8)),(np.std(p16)),(np.std(p32)),(np.std(p64)),(np.std(p128))])        
y = np.array([(np.mean(p4)),(np.mean(p8)),(np.mean(p16)),(np.mean(p32)),(np.mean(p64)),(np.mean(p128))])         

#trend = np.polyfit(x, y, 5)

#trendpoly = np.poly1d(trend) 
#plt.plot(x,trendpoly(x))
#plt.scatter(x, y)

x_new = np.linspace(0., 0.15, num=len(x)*10)
coefs = poly.polyfit(x,y , 3)
ffit = poly.polyval(x_new, coefs)

plt.scatter(x,y)
plt.plot(x_new,ffit, label=r'$p_{c}(\infty)=0.592$')
plt.xlabel(r'$\sigma $',fontsize=15.)
plt.ylabel(r'$p_{c}$',fontsize=15.)


print (poly.polyval(0, coefs))

plt.legend(fontsize=15.,frameon=False)

plt.savefig('/home/pedro/Escritorio/Fisica-Computacional-master/Corridas/extrapolacionA.pdf', bbox_inches='tight', rasterized=True, dpi=300) 
```
