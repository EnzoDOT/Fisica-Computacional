```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import ascii
import matplotlib.colors as colors
from matplotlib import *
import sys
from pylab import *
from matplotlib.colors import LogNorm
import scipy.integrate as integrate
import scipy
from astropy.io import fits
import math
import os,glob
import matplotlib.patches as mpatches
from scipy.interpolate import splrep, splev
import numpy.polynomial.polynomial as poly



N=5
mylist  = []


sigma=np.zeros((N,), dtype=np.float)
pc=np.zeros((N,), dtype=np.float)


with open('/home/pedro/Escritorio/dispersionB.dat', 'r') as Gf:
    for line in Gf:
        mylist.append(line)
Gf.close()


for i in range(N): 
        lista = mylist[i].split()
        sigma[i] = float(lista[0])
        pc[i] = float(lista[1])
        
x = sigma  
y = pc
#trend = np.polyfit(x, y, 5)

#trendpoly = np.poly1d(trend) 
#plt.plot(x,trendpoly(x))
#plt.scatter(x, y)

x_new = np.linspace(0., 0.15, num=len(x)*10)
coefs = poly.polyfit(x,y , 2)
ffit = poly.polyval(x_new, coefs)
axes = plt.gca()

plt.scatter(x,y)
plt.plot(x_new,ffit, label=r'$p_{c}(\infty)=0.596$')
plt.ylim(0.4,0.7)

print (poly.polyval(0, coefs))

plt.legend(fontsize=15.,frameon=False)
plt.xlabel(r'$\sigma$',fontsize=15.)
plt.ylabel(r'$p_{c}$',fontsize=15.)

plt.savefig('/home/pedro/Escritorio/extrapolacionB.pdf', bbox_inches='tight', rasterized=True, dpi=300) 
```
